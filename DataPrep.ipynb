### data load and processing

```python
asset_type = 'stocks' # can be one of ['stocks', 'fx', 'crypto', 'commodities']
```

```python
for f in tqdm(os.listdir('data_pkl')):
    data_timeperiod = f.replace('.pkl','').split('_')[-1]
    symbol = f.replace('.pkl','').split('_')[0]
    print(symbol, data_timeperiod)
    data_timeperiod_time = datetime.timedelta(minutes=int(data_timeperiod.replace('min','')))
    data = pkl.load(open('data_pkl/'+f, 'rb')) #get_data('SPY', period=data_timeperiod)
    try:
#         pkl.dump(data, open('datapkl/'+(f.replace('.pkl','')+'.pkl'), 'wb')) # write
        if asset_type not in ['fx', 'crypto']:
            data = data.between_time('09:30', '16:00')
        data = procdata(data)
        data = data.rename({'X__Open': 'Open',
                            'X__High': 'High',
                            'X__Low': 'Low',
                            'X__Close': 'Close',
                            'X__Volume': 'Volume',
                            }, axis=1)
        pkl.dump(data, open('datapkl/'+(f.replace('.pkl','')+'_proc.pkl'), 'wb'))
    except Exception as ex:
        print(str(ex))
```
